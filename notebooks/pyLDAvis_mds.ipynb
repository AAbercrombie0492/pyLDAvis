```python
import pyLDAvis
import pyLDAvis.sklearn
from pyLDAvis._prepare import (js_PCoA, js_MDS, js_NMDS, js_TSNE)
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
pyLDAvis.enable_notebook()
```

## load data

```python
newsgroup = fetch_20newsgroups(remove=('headers', 'footers', 'quotes'))
docs_raw = newsgroup.data
print len(docs_raw)
```

## document-term matrix

```python
tfidf_vectorizer = TfidfVectorizer(strip_accents = 'unicode', 
                                   stop_words = 'english', 
                                   token_pattern = r'\b[a-zA-Z]{2,}\b',
                                   max_df = 0.5, 
                                   min_df = 10)
```

```python
dtm_tfidf = tfidf_vectorizer.fit_transform(docs_raw)
print dtm_tfidf.shape
```

## latent dirichlet allocation

```python
lda = LatentDirichletAllocation(n_topics=20, learning_method='batch', random_state=0)
lda.fit(dtm_tfidf)
```

## pyLDAvis

### PCoa

```python
pyLDAvis.sklearn.prepare(lda, dtm_tfidf, tfidf_vectorizer, sort_topics=False, mds=js_PCoA)
```

### MDS

```python
pyLDAvis.sklearn.prepare(lda, dtm_tfidf, tfidf_vectorizer, sort_topics=False, mds=js_MDS)
```

### NMDS

```python
pyLDAvis.sklearn.prepare(lda, dtm_tfidf, tfidf_vectorizer, sort_topics=False, mds=js_NMDS)
```

### TSNE

```python
pyLDAvis.sklearn.prepare(lda, dtm_tfidf, tfidf_vectorizer, sort_topics=False, mds=js_TSNE)
```
